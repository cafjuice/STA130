## Question 2:

Accuracy is a general measure of how often a classification model correctly categorises your data, which is a useful metric in situations where the cost of false positives and false negatives are roughly equal; for example, accuracy might help evaluate a model which predicts whether a customer will click on a pop-up.

Sensitivity specifically measures how well a classification model identifies positive data, which is especially useful in situations where we need to minimise false negatives; for instance, sensitivity becomes very important in models that try to predict whether or not patients have a certain disease, since failing to identify a sick patient would have severe consequences.

Specificity is sensitivity's opposite: it measures how well a classification model identifies negative data, which is useful for when we want to minimise false positives. A (somewhat) lighthearted example might be cases where your email sorting model falsely identifies legitimate emails as spam ones, thus automatically deleting them.

Precision measures the proportion of true positive predictions, which is helpful in scenarios where positive predictions need to be reliable; for example, a model that identifies credit card fraud needs to be highly precise, since investigating a false positive might be a waste of time and money.

````python
def calculate_metrics(conf_matrix):
    TN, FP, FN, TP = conf_matrix.ravel()  # Flatten the confusion matrix into 4 values
    
    accuracy = (TP + TN) / (TP + TN + FP + FN)
    sensitivity = TP / (TP + FN)
    specificity = TN / (TN + FP)
    
    return accuracy, sensitivity, specificity

# Calculate metrics for clf confusion matrix
accuracy_clf, sensitivity_clf, specificity_clf = calculate_metrics(cm_clf)
print(f"Confusion Matrix for clf:")
print(f"Accuracy: {accuracy_clf:.3f}")
print(f"Sensitivity: {sensitivity_clf:.3f}")
print(f"Specificity: {specificity_clf:.3f}")

# Calculate metrics for clf2 confusion matrix
accuracy_clf2, sensitivity_clf2, specificity_clf2 = calculate_metrics(cm_clf2)
print(f"\nConfusion Matrix for clf2:")
print(f"Accuracy: {accuracy_clf2:.3f}")
print(f"Sensitivity: {sensitivity_clf2:.3f}")
print(f"Specificity: {specificity_clf2:.3f}")
````
